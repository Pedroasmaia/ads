# Modelagem de casos de usos

- Diagrama mais simples da UML

## Representação grafica:



- Elipse: Deve conter nome (verbo + substantivo)
- Ator: (Usuário na imagem) representa todas as pessoas e objetos que utilizam o sistema
- linha continua (a): Relacionamento entre caso de uso e ator
- linha continua com seta (b): Generalização(Herança) 2 ou mais casos de uso com a mesma caracteristica
- linha tracejada com seta (c):  Pode representar o include (relacionamento obrigatorio entre casos de uso) extend (relacionamento opcional entre casos de uso).

# Diagrama de Classes:

- Diagrama mais utilizados das nossas classes

## Representação grafica:



- Nome
- Atributos: São as variaveis;
- Métodos: Funcionalidades;
- Objeto: Classe instanciada;

### Relacionamento do tipo encapsulamento

- Public (+): Livre para qualquer classe manusear atributos.
- Private (-): Apenas a classe permitida tem acesso aos atributos.
- Protected (#): As classes filhas podem manusear os atributos.

### Outros tipos de classes



- Abstrata:
  - Nome em Italio
  - Sem instancias
  - Usada para definir hierarquias
- Interface:
  - Contem o esteriotipo "<<Name>>"
  - Representam conjunto de operaçoes que são implementadas por outras



- Gravador de som é subclasse de Aparelho de Som

### Relacionamento de multiplicidade



- n..m: significa que temos n para m instâncias entre classes;
- n..*: se utiliza o asterisco quando indica-se muitos;
- 1: exatamente uma instância associada;
- 1..* - Uma ou mais intâncias que podem estar relacionadas.
