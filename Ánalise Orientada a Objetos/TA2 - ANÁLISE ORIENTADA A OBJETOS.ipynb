- linha tracejada com seta (c):  Pode representar o include (relacionamento obrigatorio entre casos de uso) extend (relacionamento opcional entre casos de uso).

# Diagrama de Classes:

- Diagrama mais utilizados das nossas classes

## Representao grafica:



- Nome
- Atributos: So as variaveis;
- Mtodos: Funcionalidades;
- Objeto: Classe instanciada;

### Relacionamento do tipo encapsulamento

- Public (+): Livre para qualquer classe manusear atributos.
- Private (-): Apenas a classe permitida tem acesso aos atributos.
- Protected (#): As classes filhas podem manusear os atributos.

### Outros tipos de classes



- Abstrata:
  - Nome em Italio
  - Sem instancias
  - Usada para definir hierarquias
- Interface:
  - Contem o esteriotipo "<<Name>>"
  - Representam conjunto de operaoes que so implementadas por outras



- Gravador de som  subclasse de Aparelho de Som

### Relacionamento de multiplicidade



- n..m: significa que temos n para m instncias entre classes;
- n..*: se utiliza o asterisco quando indica-se muitos;
- 1: exatamente uma instncia associada;
- 1..* - Uma ou mais intncias que podem estar relacionadas.
