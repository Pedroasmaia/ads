# Análise Orientada a Objetos

## Paradigma Orientado a Objetos

- Relacionamento entre classes e objetos e o relacionamento entre eles:
  - Herança;
  - Polimorfismo;
  - Agregação;
  - Composição;

- Abstrair do mundo real para o mundo computacional.
- Classe é um conjunto de valores que podem ser transformadas em objetos



- Toda receita tem ingredientes, e um modo de fazer.
- Os programas tem atributos e métodos.

- **Classe:** Define o comportamento dos objetos através de métodos e os estados possíveis destes objetos através de atributos
- **Objeto:** Instância de uma classe
- **Herança:** As classes compartilham atributos, métodos e outros membros da classe entre sí.
- **Polimorfismo:** Métodos que têm a mesma assinatura, mas comportamentos distintos
- **Encapsulamento** Proibição do acesso direto ao estado do objeto

## Características da UML

- Combina os conceitos comuns de linguagens OO;
- Abrange desde o requisito até as etapas finais do desenvolvimento;
- Compatível com diversos escopos.

- Capturar e definir com precisão os requisitos de software;

- Permitir o fácil entendmiento de projetos complexos.

### Fluxo de desenvolvimento

- Funciona em qualquer fluxo, sendo ágil ou não;
- Apresentação visual da semântica do sistema;
- Contexto.

# Mecanismos comuns da UML

## Especificação

- Descrição exata do elemento

- No diagrama de classe existe uma especificação;
  - Atributos;
  - Operações;
  - Comportamentos;

- Adorno é a notação grafica dos modelos.

## Divisões comuns

- Divisão de classes: Sendo a classe uma abstração e o objeto uma manifestação concreta (instancia) dessa abstração;
- Separação de interface e implementação: Interface declara um contrato, e a implementação sua realização

## Mecanismo de extensão

- Estréotipos: É possivel utilizar o "desenho" de determinado bloco, e modifica-lo com algum objetivo, criando um novo objeto;
- Restrições: É possivel alterar as restrições de um diagrama;
- Valores predefinidos: É possivel predefinir valores em um diagrama.

## Regras de consistencias da UML

- O número de objetos no diagrama de sequencia deve ser o mesmo do número de classes;
- Atualizar o diagrama de sequencias conforme o diagrama de classes;
- O nome dos métodos devem ser respeitados entre os diagramas;
- O diagrama de classes e de sequencias devem ser sincronizados;
- Cada situação no caso de uso deve ter uma operação correspondente no diagrama de classes;
- Cada caso de uso deve ter um substantivo e um verbo associado;
- Para cada caso de uso deve haver um diagrama de sequencia;
- Deve haver consistencia entre os atores do diagrama de caso de uso e o de sequencia.
