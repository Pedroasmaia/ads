# Anlise Orientada a Objetos

## Paradigma Orientado a Objetos

- Relacionamento entre classes e objetos e o relacionamento entre eles:
  - Herana;
  - Polimorfismo;
  - Agregao;
  - Composio;

- Abstrair do mundo real para o mundo computacional.
- Classe  um conjunto de valores que podem ser transformadas em objetos



- Toda receita tem ingredientes, e um modo de fazer.
- Os programas tem atributos e mtodos.

- **Classe:** Define o comportamento dos objetos atravs de mtodos e os estados possveis destes objetos atravs de atributos
- **Objeto:** Instncia de uma classe
- **Herana:** As classes compartilham atributos, mtodos e outros membros da classe entre s.
- **Polimorfismo:** Mtodos que tm a mesma assinatura, mas comportamentos distintos
- **Encapsulamento** Proibio do acesso direto ao estado do objeto

# Mecanismos comuns da UML

## Especificao

- Descrio exata do elemento

- No diagrama de classe existe uma especificao;
  - Atributos;
  - Operaes;
  - Comportamentos;

- Adorno  a notao grafica dos modelos.

## Divises comuns

- Diviso de classes: Sendo a classe uma abstrao e o objeto uma manifestao concreta (instancia) dessa abstrao;
- Separao de interface e implementao: Interface declara um contrato, e a implementao sua realizao

## Mecanismo de extenso

- Estrotipos:  possivel utilizar o "desenho" de determinado bloco, e modifica-lo com algum objetivo, criando um novo objeto;
- Restries:  possivel alterar as restries de um diagrama;
- Valores predefinidos:  possivel predefinir valores em um diagrama.

## Regras de consistencias da UML

- O nmero de objetos no diagrama de sequencia deve ser o mesmo do nmero de classes;
- Atualizar o diagrama de sequencias conforme o diagrama de classes;
- O nome dos mtodos devem ser respeitados entre os diagramas;
- O diagrama de classes e de sequencias devem ser sincronizados;
- Cada situao no caso de uso deve ter uma operao correspondente no diagrama de classes;
- Cada caso de uso deve ter um substantivo e um verbo associado;
- Para cada caso de uso deve haver um diagrama de sequencia;
- Deve haver consistencia entre os atores do diagrama de caso de uso e o de sequencia.
