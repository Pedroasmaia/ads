# Lógica e Matemática Computacional

## Fundamentos da Lógica

### Lógica:

- É a arte de pensar corretamente;
- "Ciência das formas do pensamento" - Forbellone;
- Lógica nos ajuda a pensar de forma organizada;
- Lógica é a ciência que estuda os argumentos, premissas e conclusões;
- **Definições:**
    - **Proposição:** Um enunciado, frase declarativa, sendo possível declarar ser verdadeiro ou falso.
    - **Premissas:** Proposição base para um raciocínio.
    - **Argumento:** Conjunto de preposições relacionadas.
    - **Silogismo:** Raciocínio dedutivo que permite dedução de uma conclusão a partir de uma premissa.
    - **Falácia:** Argumentos ilógicos.
    - **Inferência:** Processo que permite chegar a conclusões a partir de premissas, podendo ser indutiva e dedutiva. Uma inferência falsa é a falácia.

### Lógica formal:

- Começa com Aristóteles,
- Lida com relações entre premissas e conclusões,

- **Indutiva:** Experiência com a verdade e fatos particulares para chegar a uma conclusão;
    - Combinações e Possibilidades.
    - O Sol nasceu todas as manhãs até hoje. Logo é provável que nasça amanha.
- **Dedutiva:** Premissas gerais para concluir verdades especificas e particulares.
    - Clássica
    - Todo homem é mortal. Pedro é um homem, logo Pedro é mortal.

## Evolução da lógica

- Podemos demonstrar a lógica por meio da matemática;
- A lógica pode ser separada em 3 grandes períodos:
    - Aristotélico
    - Booleano
    - Atual
- Lógica se preocupa com a construção do pensamento. Assumindo a premissa verdadeira.

### Período Aristotélico Lógica Clássica:

- Identidade: Uma proposição verdadeira é verdadeira, Uma proposição falsa é falsa.
- Não contradição: Uma proposição não pode ser verdadeira e falsa ao mesmo tempo.
- Terceiro excluído: Toda proposição é verdadeira ou é falsa.

### Álgebra Booleana

- Inventado por George Boole.
- Primeiro sistema detalhado a usar a lógica como cálculo.
- A adição é associada ao conectivo *ou*
    (A)=1
    (B)=0
    A+B=1(Verdadeiro)
- A multiplicação é associado ao conectivo *e*
    (A)=1
    (B)=0
    A*B=0(Falso)

### Álgebra dos Conjuntos

- Inventado por Georg Cantor
- Adivinda das teorias dos conjuntos
- Operadores lógicos:
    - **~**: Negação
    - **^**: Conjunção(and/e)
    - **v**: Disjunção(or/ou)
    - **→**: Condicional(if/se,então)
    - **←**: Bicondicional(se,somente se)

## Princípios matemáticos

- A matemática discreta (matemática combinatória) é um ramo da matemática voltado ao estudo de objetos finitos
- Usada quando contamos objetos, processos envolvendo um número finitos.
    - Problemas de existência.
    - Problemas de contagem.
    - Problemas de otimização.
