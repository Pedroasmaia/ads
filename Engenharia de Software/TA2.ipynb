### Levantamento de Requisitos

- **Funcionais:** Apresentam novas funções ao sistema.
    - Exemplos:
        - Gerar Relatorios 
        - Incluir/Excluir/Alterar cadastros
- **Não funcionais:** São caracteristicas do sistema.
    - Exemplos:
        - Deverá executar em Windows e Linux
        - Se Comunicar com SQL Server
        - Página carregue em 5 milisegundos

## Qualidade no Processo

### CMMI

- É um modelo que aborda a melhoria de processos.

- Divido em 5 Níveis de maturidade:

![Alt text](image-8.png)

- 3 Modelos:
 - CMMI For Development
 - CMMI For Acquisition
 - CMMI For Services

- Contém 22 áreas de processos que são divididas em 4 grupos:

    - Gerenciamento de Processos
        - Foco no processo organizacional
        - Definição do processo
        - Treinamento
        - Desempenho do processo organizacional
    - Gerenciamento de Projetos
        - Planejamento dos projetos
        - Monitoração e controle de projeto
        - Fornecedores
        - Gerenciarmento de Risco
        - Gerenciamento de Projetos Integrados
    - Engenharia
        - Gerenciamento de requisitos
        - Desenvolvimento de Requisitos
        - Solução Técnica
        - Integração de produto
        - Verificação
        - Validação
    - Apoio
        - Gerenciamento de configuração
        - Gerenciamento de qualidade de processo e produto
        - Medição e análise
        - Analise de decisão e resolução
        - Analise casual e resolução

## Medição de Qualidade

- **Falha de Software:** Comportamento inesperado do sistema;
- **Erro**: Execuções incorretas;
- **Defeitos**: Implementação Incorreta
- **Bug**: Erros e falhas inesperados

- **Medidas:** indicação quantitativa da extensão, quantidade, capacidade ou tamanho de um atributo,produto ou processo
- **Métricas:** medida quantitativa do grau com qual um sistema, componente ou processo possui determinado atribudo
- **Indicador** uma metrica ou varias dela que fornecem informação sobre o processo de software
