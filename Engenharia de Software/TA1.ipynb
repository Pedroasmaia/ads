# Fundamentos da Engenharia de Software

## Princpios da Engenharia de Software

- Organizao hierrquica;
- Formalidade;
- Completeza;
- Dividir para conquistar;
- Ocultao;
- Localizao
- Integridade conceitual;
- Abstrao;

### Fundamentos:

"Engenharia de software  uma disciplina com objetivo de produzir software sem falhas, entregue dentro do prazo e do oramento, que atenda as necessidades do cliente e de facil manuteno."

### Tipos de Software:

- Sistemas: Operacionais, do suporte para outros softwares;
- Aplicaes: Interagem com o usurio;
- Engenharia/cientfico;
- Embarcado; Dentro de Hardwares
- Web;
- Aplicativos mveis;
- Inteligncia artificial

### Crise de Software:

- Baixa qualidade
- Prazos e Oramento estourados
- Dificil de manter

### Eternos Mitos:

- Manual de procedimentos produz qualidade.
- Mais programadores aceleram o processo.
- Requisitos mudam o tempo todo, e sempre  possvel acomod-los.
- Testes podem remover todos os erros.

### Modelo de processo:

- Representao dos objetos e atividades
-  uma forma de montar etapas para serem seguidas, "um fluxograma"

### Processo de Software

- Sequncia de passos executados com um determidado objetivo

#### Atividades Fundamentais:

- Especificao de software: Validar requisitos
- Projeto e Implantao de Software: Projeto de Arquitetura
- Validao de Software: Testes unitrios
- Evoluo de software: Projeto de Manunteno de Software

#### Modelo Cascata:



 um mtodo linear, e  antiquado, mas ainda serve em alguns softwares embarcados

#### Modelo Espiral



Pequenas variaes com pequenas partes sendo entregues.


#### Modelo V



Focado na qualidade. Sequencial envolvendo testes em todas as etapas.

#### Modelo Incremental



Combina fluxo de processos lineares, de forma paralelas. Sempre incrementando modificaes.


## Metodologias Ageis

- Comeou em 1990 como reao aos mtodos pesados.
- Menos centradas em documentao.
- So adaptativas em vez de determinantes.
- Orientado a pessoas e no a processos.
- Planejamento incrimental e iterativa.
- Mais envolvimento do cliente.


### Extreme Programming (xp)

- Rapido desenvolvimento
- Requisitos se alteram constantemente
- Valores do XP:
    - Comunicao
    - Simplicidade
    - Feedback
    - Coragem

- Equipe de Trabalho:
    - Gerente de Projeto: Trabalhos administrativos, incluindo o Cliente
    - Coach: Responsvel tcnico pelo Projeto
    - Analista de Teste: Ajuda o cliente a escrever os testes
    - Redator Tcnico: Responsvel pela Documentao
    - Desenvolvedor

### Scrum:

- Focado na gesto de projetos, tem como base o planehamento iterativo e incrimental
- Reunio Regulares:
    - Daily (Rapida, para dizer o que foi feito)
- Projetos so dividos em ciclos (sprints)
    - Tempo determinado.
- entrega incrimental
- Envolvimento do cliente

![Alt text](image-1.png)

-Equipe de Trabalho:
    - Scrum Master: Facilitador, tirar os impedimentos do time
    - Product Owner: Pessoa responsvel pelo projeto
    - Scrum Team: Equipe de Desenvolvimento
- Elementos:
    - Product Backlog: Trata-se da lista de todas as funcionalidades desejadas para o produto
    - Sprint Backlog: Lista de tarefas que so realizadas na Sprint
    - Sprint: Ciclos regulares de duas a quatro semanas.
    - Kanban: Quadro para organizar as tarefas.
    - Poquer de Planejamento:  utilizado para escolher os itens que sero feitos, e quanto tempo demorara para fazer cada item.

### Quadro do Scrum

- Meio visual de gerir as atividades do projeto.
- Oficialmente, no faz parte da metodologia agil, mas  usado de forma ampla.
- Divises do quadro:
    - estria: Relato dos requisistos
    - ToDo (A Fazer)
    - In progress (Em progresso)
    - Done (Feito)

### Elementos importantes:

- Repositrio
- Baselines (Linhas de base) | Releases
- Branches | Ramificaes dos repositrios
- Item de configuraes, qualquer componente que precise de controle de verso:
    - Codigos,
    - Desenhos,
    - Testes e outros;

 importante para manter uma rastreabilidade, devem ter nomes hieraquizados.


### Banco de dados de configuraes

- Registrar todas as informaes relevantes sobre a configuraes do sistestma e os itens de configurao;
- Avaliar o impacto das mudanas no sistema
- Registro e recuperao de informaes de projeto

### Gerenciamento de Verses e Releses:

-  para a indentificao e rastreabilidade
- Relesases: so versoes distribuidas aos clientes
    - Tem funcionalidades diferentes
    - Sempre temos mais verses que releases
