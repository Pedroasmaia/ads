# Fundamentos da Engenharia de Software

## Princípios da Engenharia de Software

- Organização hierárquica;
- Formalidade;
- Completeza;
- Dividir para conquistar;
- Ocultação;
- Localização
- Integridade conceitual;
- Abstração;

### Fundamentos:

"Engenharia de software é uma disciplina com objetivo de produzir software sem falhas, entregue dentro do prazo e do orçamento, que atenda as necessidades do cliente e de facil manutenção."

### Tipos de Software:

- Sistemas: Operacionais, dão suporte para outros softwares;
- Aplicações: Interagem com o usuário;
- Engenharia/científico;
- Embarcado; Dentro de Hardwares
- Web;
- Aplicativos móveis;
- Inteligência artificial

### Crise de Software:

- Baixa qualidade
- Prazos e Orçamento estourados
- Dificil de manter

### Eternos Mitos:

- Manual de procedimentos produz qualidade.
- Mais programadores aceleram o processo.
- Requisitos mudam o tempo todo, e sempre é possível acomodá-los.
- Testes podem remover todos os erros.

### Modelo de processo:

- Representação dos objetos e atividades
- É uma forma de montar etapas para serem seguidas, "um fluxograma"

### Processo de Software

- Sequência de passos executados com um determidado objetivo

#### Atividades Fundamentais:

- Especificação de software: Validar requisitos
- Projeto e Implantação de Software: Projeto de Arquitetura
- Validação de Software: Testes unitários
- Evolução de software: Projeto de Manuntenção de Software

#### Modelo Cascata:



É um método linear, e é antiquado, mas ainda serve em alguns softwares embarcados

#### Modelo Espiral



Pequenas variações com pequenas partes sendo entregues.


#### Modelo V



Focado na qualidade. Sequencial envolvendo testes em todas as etapas.

#### Modelo Incremental



Combina fluxo de processos lineares, de forma paralelas. Sempre incrementando modificações.


## Metodologias Ageis

- Começou em 1990 como reação aos métodos pesados.
- Menos centradas em documentação.
- São adaptativas em vez de determinantes.
- Orientado a pessoas e não a processos.
- Planejamento incrimental e iterativa.
- Mais envolvimento do cliente.


### Extreme Programming (xp)

- Rapido desenvolvimento
- Requisitos se alteram constantemente
- Valores do XP:
    - Comunicação
    - Simplicidade
    - Feedback
    - Coragem

- Equipe de Trabalho:
    - Gerente de Projeto: Trabalhos administrativos, incluindo o Cliente
    - Coach: Responsável técnico pelo Projeto
    - Analista de Teste: Ajuda o cliente a escrever os testes
    - Redator Técnico: Responsável pela Documentação
    - Desenvolvedor

### Scrum:

- Focado na gestão de projetos, tem como base o planehamento iterativo e incrimental
- Reunião Regulares:
    - Daily (Rapida, para dizer o que foi feito)
- Projetos são dividos em ciclos (sprints)
    - Tempo determinado.
- entrega incrimental
- Envolvimento do cliente

![Alt text](image-1.png)

-Equipe de Trabalho:
    - Scrum Master: Facilitador, tirar os impedimentos do time
    - Product Owner: Pessoa responsável pelo projeto
    - Scrum Team: Equipe de Desenvolvimento
- Elementos:
    - Product Backlog: Trata-se da lista de todas as funcionalidades desejadas para o produto
    - Sprint Backlog: Lista de tarefas que são realizadas na Sprint
    - Sprint: Ciclos regulares de duas a quatro semanas.
    - Kanban: Quadro para organizar as tarefas.
    - Poquer de Planejamento: É utilizado para escolher os itens que serão feitos, e quanto tempo demorara para fazer cada item.

### Quadro do Scrum

- Meio visual de gerir as atividades do projeto.
- Oficialmente, não faz parte da metodologia agil, mas é usado de forma ampla.
- Divisões do quadro:
    - estória: Relato dos requisistos
    - ToDo (A Fazer)
    - In progress (Em progresso)
    - Done (Feito)

## Gestão de Comunicação

- Definir o que será gerenciado e o esquema que será usado para identificar as entidades.
- Estabelecer o responável
- Estabelecer políticas de gerenciamento de configuração.
- Especificar de ferramentas.
- Descrever a estrutura do banco de dados.

![Alt text](image-4.png)


### Elementos importantes:

- Repositório
- Baselines (Linhas de base) | Releases
- Branches | Ramificações dos repositórios
- Item de configurações, qualquer componente que precise de controle de versão:
    - Codigos,
    - Desenhos,
    - Testes e outros;

É importante para manter uma rastreabilidade, devem ter nomes hieraquizados.


### Banco de dados de configurações

- Registrar todas as informações relevantes sobre a configurações do sistestma e os itens de configuração;
- Avaliar o impacto das mudanças no sistema
- Registro e recuperação de informações de projeto

### Gerenciamento de Versões e Releses:

- É para a indentificação e rastreabilidade
- Relesases: são versoes distribuidas aos clientes
    - Tem funcionalidades diferentes
    - Sempre temos mais versões que releases
