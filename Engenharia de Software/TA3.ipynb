# Testes de Software

## Verificação e Validação de Testes

- Processo de verificação e análise;
- Visa estabelecer a confiança que o sistema de software está adequado ao seu propósito.
- Ocorre em cada estágio do processo do software:
    - Revisão de requisitos;
    - Revisões no projetos
    - Inspeções no código
    - Testes no produto

### Verificação (Desenvolvimento)

- Verificar se o software está de acordo com suas especificações
- Verificar se o software atende aos requisitos

| Estamos produzindo o produto corretamente?

### Validação (Negocio)

- Processo
- Assegurar se o produto atende às expectativas do cliente

| Estamos produzindo o produto correto?

### Teste de Software

- Sequencia de ações executadas com objetivo de encontrar problemas no software
- Busca encontrar defeitos e não perfeição
- 4 Etapas:
    - Planejamento
    - Projeto de casos de teste
    - Execução de programas com os casos de testes
    - Análise dos resultados

#### Casos de teste:

- Entrada no programa e a saída correspondente
- Deve-se escolher um bom caso de teste
- Casos de teste com baixa qualidade não exercitam partes criticas do programa

#### Planejamento dos casos de teste:

- Definir o ambiente no qual o teste sera realizado
- Definir as entradas deste caso de teste
- Definir as saida para cada entrada
- Definir os passos para serem realizados em cada teste

#### Resultados possiveis em cada teste:

- Passou: Todos os passos foram concluidos com sucesso
- Falhou: Nem todos os passos foram obtiveram sucesso
- Bloqueado: O teste não pode ser realizado

#### Defeito, Falha ou Erro

- Defeito é falha de lógica na programação
- Falha é o não funcionamento de um programa por causa de um defeito
- Erro é o resultado de um processamento não retorna como o esperado devido a um defeito

### Teste de unidade

- Direcionada a uma pequena parte do produto, rotina , classes.
- Quem executa normalmente é o desenvolvedor
- Stub: é um trecho que vai substituir entradas para testar o código

### Teste de integração

- Integração Top-down: Desenvolve primeiro o esqueleto do sistema e depois preenche com componentes

- Integra bottom-up: Integra primeiro os componentes de infraestrutura e depois adiciona os componentes funcionais.
![Alt text](image-13.png)

### Testes de validação

- Começa quando termina os testes de integração, quando os componentes individuais já foram executados.
- Foca em  ações visíveis ao usuário e saídas do sistema reconhecíveis pelo usuário.
- O software deve funcionar de maneira razoavelmente esperado pelo sistema.

### Teste de sistema

- Testa a aviabilidade do sistema
- Se Conversa bem com o Hardware
- Funciona bem com outros softwares 

## Técnicas de teste

### Testes Funcionais

- Teste de caixa preta
- O Código fonte é ignorado
- Baseado nos requisitos básicos
- Preocupado em como o software funciona

- 2 Passos:
    - Identificar funções, requisitos
    - Criar casos de testes

- Utilizado para:
    - Funções incorretas e Omitidas
    - Erros de comportamento
    - Erros de desempenho
    - Erros de interface
    - Erros de iniciação e termino

### Testes Estruturais

- Teste de caixa branca
- São derivados do conhecimento da estrutura e da implementação do software
- Conhecendo a estrutura de software, auxilia na identificação de partições e casos de testes adicionais

### Teste de Desempenho

- Objetivos:
    - Avaliar disponibilidade
    - Analisar criterios de desempenho
    - Comparar as caracteristicas de desempenho de vários sistemas ou configurações
    - Procurar a fonte de um problema
    - Encontrar os níveis de rendimento

### Teste de Estresse 

- Funções:
    - Testar o comportamento de falha do sistema: Testa se a sobrecarga acarreta uma falha leve em vez de um colapso do sistema
    - Estressar o sistema e analisar o surgimento de defeitos
