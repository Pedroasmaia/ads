# Auditoria de Sistema

## Fundametos de Auditoria de Sistema

Auditoria: Principal Funo analisar parcial ou globalmente os processos, resultados e sugerir aes de melhoria ou corretivas
- No obriga aes
- Pautada por Testes
- Onde pode ser aplicado:
  - Processos: Como as coisas so feitas.
  - Desenvolvimentos: Analise do cdigo.
  - Testes: Se os testes esto sendo bem feitos.
  - Segurana e proteo de dados: Faz tratativa em cima da segurana.
  - Estrutura de desenvolvimento: Observa mais a administrao da Empresa.

### Auditoria do Sistema da informao

- Verifica toda a estrutura computacional
- Eficincia dos Processos
- Segurana dos dados
- Cumprimento da Lei

#### Abordagens:

- Ao Redor do computador:
  - Confronta os documentos fontes com o resultado
  - Utiliza-se rotinas manuais
  - Sua aplicao envolve custos baixos e diretos
  - No exige muito conhecimento sobre Tecnologia da Informao.

- Atravs do computador: 
  - Capacita o auditor a fazer verificaes constantes.
  - Faz aprovao de LOGS armazenados
  - No deixa evidencias documentais
  - Se for efetuada incorretamente tras grandes perdas

- Com o computador:
  - Mais efetuada e mais completa
  - Utiliza as capacidades do computador
  - Podemos desenvolver programas especificos para serem usados pelo auditor
  - Faz processos manuais e automatizados


#### Auditor

- Inspeo: Inspecionar e Verificar onde ocorrem as inconformidades e apresentar provas.
- Controle: Efetuar analise do controle dos processos.
- Risco: Apresentar os riscos podem afetar o projeto.
- Contnua: Analiser diversos momentos.
