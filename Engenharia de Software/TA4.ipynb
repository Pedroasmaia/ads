# Auditoria de Sistema

## Fundametos de Auditoria de Sistema

Auditoria: Principal Função analisar parcial ou globalmente os processos, resultados e sugerir ações de melhoria ou corretivas
- Não obriga ações
- Pautada por Testes
- Onde pode ser aplicado:
  - Processos: Como as coisas são feitas.
  - Desenvolvimentos: Analise do código.
  - Testes: Se os testes estão sendo bem feitos.
  - Segurança e proteção de dados: Faz tratativa em cima da segurança.
  - Estrutura de desenvolvimento: Observa mais a administração da Empresa.

### Auditoria do Sistema da informação

- Verifica toda a estrutura computacional
- Eficiência dos Processos
- Segurança dos dados
- Cumprimento da Lei

#### Abordagens:

- Ao Redor do computador:
  - Confronta os documentos fontes com o resultado
  - Utiliza-se rotinas manuais
  - Sua aplicação envolve custos baixos e diretos
  - Não exige muito conhecimento sobre Tecnologia da Informação.

- Através do computador: 
  - Capacita o auditor a fazer verificações constantes.
  - Faz aprovação de LOGS armazenados
  - Não deixa evidencias documentais
  - Se for efetuada incorretamente tras grandes perdas

- Com o computador:
  - Mais efetuada e mais completa
  - Utiliza as capacidades do computador
  - Podemos desenvolver programas especificos para serem usados pelo auditor
  - Faz processos manuais e automatizados


#### Auditor

- Inspeção: Inspecionar e Verificar onde ocorrem as inconformidades e apresentar provas.
- Controle: Efetuar analise do controle dos processos.
- Risco: Apresentar os riscos podem afetar o projeto.
- Contínua: Analiser diversos momentos.
