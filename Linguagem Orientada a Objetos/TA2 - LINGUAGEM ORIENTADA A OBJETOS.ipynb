## Estrutura de decisão

- IF e IF-ELSE



```python
num = int(input("Insira seu número: ")) 
# Se número for >(maior que) 10 
if num > 10:
  #Mostra esse texto:
  print("O Número é grande")
else:
  #Caso contrario
  print("O Número é pequeno")
```

### Operador Ternario

Diminuição de linhas para escrever if e else


```python
print("Grande" if num > 10 else "Pequeno")
```
