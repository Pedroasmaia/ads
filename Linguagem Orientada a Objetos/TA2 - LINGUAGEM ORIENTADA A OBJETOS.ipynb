## Estrutura de deciso

- IF e IF-ELSE



```python
num = int(input("Insira seu nmero: ")) 
# Se nmero for >(maior que) 10 
if num > 10:
  #Mostra esse texto:
  print("O Nmero  grande")
else:
  #Caso contrario
  print("O Nmero  pequeno")
```

### Operador Ternario

Diminuio de linhas para escrever if e else


```python
print("Grande" if num > 10 else "Pequeno")
```
