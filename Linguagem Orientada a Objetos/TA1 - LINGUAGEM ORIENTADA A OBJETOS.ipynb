# Conceitos do Paradigma Orientada a Objetos

Paradigma de programação é a forma de escrita, a estrutura de determinado código, podendo ser aceita ou não pela linguagem.


## Programação Orientada a Objetos

- Aproxima a manipulação das estrutura de um programa ao manuseio das coisas no mundo real.
- Pode ser considerado uma filosofia que guia todo o desenvolvimento.
- Principal caracteristicas:
  - Reutilizar código.

## Programação Procedural

- Se caracteriza por utilizar funções e procedimentos como núcleo de organização estrutural.
- Principal caracteristicas:
  - Utiliza código em sequencia.

## Programação Imperativo

- Declara comandos dizendo como o código deve executar algo.

## Programação Declarativa

- Descreve o que deve ser feito, mas não como deve ser feito.



# 4 Pilares da POO

1- Abstração:
  - Definir e focar no que é necessário no desenvolvimento
  - O que é necessario para o seu sistema?
  
  

2- Encapsulamento:
  - Detalhes internos do funcionamento da classe são escondidadas, impedidndo mudanças do seus atributos.

3- Herança:
  - Representa a relação entre classes diferentes que podem reutilizar código, permitindo uma herdar atributo da __clase pai__

4- Polimorfismo:
  - Diz respeito à capacidade da lingagem de processar objetos de formas diferentes dependendo do seu tipo de dado ou classe.

## Classes

- Representam algo do mundo real ou imaginário.
- É o projeto ou a modelagem de algo. (É a forma do bolo).
- Contem atributos que representam suas caracteristicas, váriaveis.



- Método dá ao objeto a capacidade de executar algum tipo de ação, comportamento ou processamento.





## Objeto

- É a instancia, ou a manifestação de uma classe.
- O Objeto é o projeto real (Cada bolo é um objeto).




# Terminologias

- IDE: Integrate Development Environment. Editor de códigos.(Exemplo: Vscode, NetBeans, Eclipse)

~~~java
  public class MyClass {
  public static void main(String args[]){
    System.out.println("Olá Mundo");
  }
}
~~~

- *public* e *class* são palavras reservadas
- *{}* são usados pra definir bloco
- *public static void* é uma função (ou método estático) publica chamada _main_ que não retorna nenhum valor
