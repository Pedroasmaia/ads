# Conceitos do Paradigma Orientada a Objetos

Paradigma de programao  a forma de escrita, a estrutura de determinado cdigo, podendo ser aceita ou no pela linguagem.


## Programao Orientada a Objetos

- Aproxima a manipulao das estrutura de um programa ao manuseio das coisas no mundo real.
- Pode ser considerado uma filosofia que guia todo o desenvolvimento.
- Principal caracteristicas:
  - Reutilizar cdigo.

## Programao Procedural

- Se caracteriza por utilizar funes e procedimentos como ncleo de organizao estrutural.
- Principal caracteristicas:
  - Utiliza cdigo em sequencia.

## Programao Imperativo

- Declara comandos dizendo como o cdigo deve executar algo.

## Programao Declarativa

- Descreve o que deve ser feito, mas no como deve ser feito.



# 4 Pilares da POO

1- Abstrao:
  - Definir e focar no que  necessrio no desenvolvimento
  - O que  necessario para o seu sistema?
  
  

2- Encapsulamento:
  - Detalhes internos do funcionamento da classe so escondidadas, impedidndo mudanas do seus atributos.

3- Herana:
  - Representa a relao entre classes diferentes que podem reutilizar cdigo, permitindo uma herdar atributo da __clase pai__

4- Polimorfismo:
  - Diz respeito  capacidade da lingagem de processar objetos de formas diferentes dependendo do seu tipo de dado ou classe.
